```python
import numpy as np
import tensorflow as tf
docs=['pakistan','go pakistan','jeetai ka bhai jeetai ga ','hip hip hurrie','baber baber']
docs
```

```python
#EMBEDDINGS INCREASE THE TEST ACCURACY OF THE DATA AS IN CASE OF TRAING IT IS OVERFITTING BUT WE WILL SEE IT LATER ON HOW TO DEAL WITH IT.
from tensorflow.keras.preprocessing.text import Tokenizer
# tokenizer=Tokenizer(oov_token='<nothing>')
tokenizer=Tokenizer()

tokenizer.fit_on_texts(docs)
len(tokenizer.word_index)

tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
sequence=tokenizer.texts_to_sequences(docs)
sequence
```

```python
from keras.utils import pad_sequences
sequences=pad_sequences(sequence,padding='post')
sequences
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
(x_train,y_train),(X_test,Y_test)=imdb.load_data()

```

```python
len(x_train[0])
```

```python
X_train = pad_sequences(x_train,padding='post',maxlen=100)
X_test = pad_sequences(X_test,padding='post',maxlen=100)
len(x_train[0])
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=10,validation_data=(X_test,Y_test))
```
